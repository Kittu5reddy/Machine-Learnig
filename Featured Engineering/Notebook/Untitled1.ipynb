```python
# Sample data with some missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen', 'Ian', 'Jack'],
    'Age': [23, 24, 22, None, 25, 24, None, 23, 22, 24],
    'Marks': [85, 90, 95, 88, None, 76, 89, None, 91, 93],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F', 'M', 'M'],
    'Passed': ['Yes', 'Yes', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes']
}
df = pd.DataFrame(data)
```

```python
df=df.fillna({"Age":df.Age.mean(),
          "Marks":df.Marks.median()})
```

```python
def generate_grades(marks):
    if marks>90 and marks<=100:
        return "O"
    elif marks>80 and marks<=90:
        return "A+"
    elif marks>70 and marks<=80:
        return "A"
    elif marks>60 and marks<=50:
        return "B+"
    else:
        return "F"
    
```

```python
generate_grades(22)
```

```python
df['Grades']=0
```

```python
df
```

```python
df['Grades']=list(map(generate_grades,df.Marks))
```
