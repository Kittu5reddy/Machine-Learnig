```python
def generate_grade(marks):
    if marks>90 and marks<=100:
        return "A+"
    elif marks>80 and marks<=90:
        return "A"
    elif marks>70 and marks<=80:
        return "B+"
    elif marks>60 and marks<=70:
        return "B"
    else:
        return "F"
```
