**Dropping and Setting Index**

```python
df = df.drop('Grade', axis=1)  # Drop column
df
```

**Cleaning Data (Handling Nulls)**

**Replace Values**

```python
df['Passed'] = df['Passed'].replace({'Yes': 1, 'No': 0})
df['Passed']
```

**GroupBy**

```python
# Average marks by gender
df.groupby('Gender')['Marks'].mean()
```

```python
# Count of students passed by Gender
df.groupby('Gender')['Passed'].sum()
```

**Duplicates**

```python
df = df.drop_duplicates()       # Remove them
df
```
